**Problem Statement**: Given above data build a machine learning model that can predict percapitaIncome based on year


Per-capita Income can be represented as following equation,

Per-capita Income = m * (year) + b

Generic form of same equation is, y=mx+b

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
new='PCI'
df['PCI']=(df['per capita income (US$)'])
df.info()
del df['per capita income (US$)']
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per-capita income')
plt.scatter(df.year,df.PCI,color='red',marker='+')
```

```python
new_df = df.drop('PCI',axis='columns')
new_df.head()
```

```python
price = df.PCI
price.head()
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

**(1) Predict Per-capita income for 2018**

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
